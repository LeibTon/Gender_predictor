```python
import requests
import re
boy = []
rex = r'(?<=<div class="col-7 col-md-3"><a href=")[^"]*'
count = 0
url = 'https://hindunames.net/hindu-boy-names-'
for i in range(1,120):
    print(i,end = "  ")
    r = requests. get(url+str(i))
    page_source = r. text
    matches = re.findall(rex,page_source)
    # print(matches)
    for i in matches:
        name = i.replace("https://hindunames.net/names/info/","")
        name = name.replace(".html","")
        name+='\n'
        boy.append(name)
        count+=1
print("Total boys name:",count)
```

```python
import requests
import re
girl = []
count = 0
rex = r'(?<=<div class="col-7 col-md-3"><a class="pink" href=")[^"]*'
url = 'https://hindunames.net/hindu-girl-names-'
for i in range(1,64):
    print(i,end = "  ")
    r = requests. get(url+str(i))
    page_source = r. text
    matches = re.findall(rex,page_source)
    # print(matches)
    for i in matches:
        name = i.replace("https://hindunames.net/names/info/","")
        name = name.replace(".html","")
        print(name)
        name+='\n'
        girl.append(name)
        count+=1
print("Total girls name:",count)
```
